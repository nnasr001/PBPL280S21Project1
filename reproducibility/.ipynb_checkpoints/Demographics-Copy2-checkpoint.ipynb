```python
ca_tracts['GEOID'] = [v[:5] for v in ca_tracts.GEOID.values]
```

```python
ca_tracts.groupby(by='GEOID').sum()
```

```python
demographics = gpd.read_parquet('dem_race_count.parquet')
```

```python
demographics.head()
```

```python
pop_dem = demographics.merge(ca_tracts.groupby(by='GEOID').sum(),on='GEOID')
```

```python
pop_dem.head()
```

```python
pop_dem['white_rate'] = pop_dem['white']/pop_dem['total_pop']
```

```python
pop_dem['black_rate'] = pop_dem['black']/pop_dem['total_pop']
```

```python
pop_dem['amind_rate'] = pop_dem['amind']/pop_dem['total_pop']
```

```python
pop_dem['asian_rate'] = pop_dem['asian']/pop_dem['total_pop']
```

```python
pop_dem['nh_pi_rate'] = pop_dem['nh_pi']/pop_dem['total_pop']
```

```python
pop_dem['two_or_more_rate'] = pop_dem['two_or_more']/pop_dem['total_pop']
```
